```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
plt.figure(figsize=(8, 5))

# Parametros de la distribucion normal
media = 70
desv_std = 4
total_estudiantes = 500

# Variables para el calculo de que los estudiantes que pesen mas de 68kg y menos de 72kg
x1 = 68
x2 = 72

# Primero necesitamos obtener la probabilidad acumulada hasta 68 kg
prob_hasta_68kg = norm.cdf(x1, media, desv_std)

# Ahora la probabilidad acumulada hasta 72kg
prob_hasta_72kg = norm.cdf(x2, media, desv_std)

# Calculamos el numero de estudiantes que pesen mas de 62kg y menos de 72kg
rango = prob_hasta_72kg - prob_hasta_68kg
estudiantes_rango = rango * total_estudiantes

print(f"\nProbabilidad de estudiantes que pesan entre 68 y 72kg: ", rango)
print("Numero de estudiante que pesan entre 68 y 72 kg: ", (estudiantes_rango))

# Graficar la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=np.logical_and(x >= x1, x <= x2), color='skyblue', alpha=0.5)
plt.title(f"Probabilidad de estudiantes que pesen entre 68 y 72 kg. \n $\mu=70$, $\sigma=4$, X=500")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad de probabilidad")
plt.grid(True)
plt.show()
```
