Ejemplo 1: La probabilidad de una vacuna

```python
from scipy.stats import binom

# Parametros de la distribucion binomial
probabilidad_exito = 0.83
num_pacientes = 16
x = 16 # Numero de exitos

#Calcular la probabilidad de que ningun aciente sufra la enfermedad
probabilidad_ningun_enfermo = binom.pmf(x, num_pacientes, probabilidad_exito)

print(f"La probabilidad de que ningun paciente sufra la enfermedad es: {probabilidad_ningun_enfermo}")
```

```python
from scipy.stats import binom

# Parametros de la distribucion binomial
probabilidad_exito = 0.83
num_pacientes = 16
x = 0

#Calcular la probabilidad de que todos los pacientes sufran la enfermedad
probabilidad_todos_enfermos = binom.pmf(x, num_pacientes, probabilidad_exito)

print(f"La probabilidad de que todos los pacientes sufran la enfermedad es: {probabilidad_todos_enfermos}")
```
