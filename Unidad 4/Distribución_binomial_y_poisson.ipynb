**Problema 1.**
## Este problema la solucion es la distribucion de poisson ya que esta describe el numero de veces que ocurre un evento en un intervalo de tiempo y con que promedio de personas ocurre

```python
from scipy.stats import poisson
# Parametros de la distribucion de Poisson
media_ingresos_dia = 5
#¿Cual es la probabilidad de que el numero de personas ingresadas
# a una unidad de cuidados intensivos en un hospital particular,
# en un dia particular, sea 2?
X = 2
probabilidad_2_ingresos = poisson.pmf(X, media_ingresos_dia)
print("1. Probabilidad de que el número de personas ingresadas sea dos:", probabilidad_2_ingresos)

# ¿Es probable que x exceda de 10?
probabilidad_10_o_menos_ingresos = poisson.cdf(10, media_ingresos_dia)
probabilidad_exceda_10 = 1 - probabilidad_10_o_menos_ingresos
print("2. Probabilidad de que el número de personas exceda de 10:", probabilidad_exceda_10)
# La probabilidad de que exceda 10 no es probable ya que lo ingresos por dia son 5
```

**Problema 2**
## De este problema la solucion es la distribucion binomial ya que tenemos el numero de ensayos, tenemos los dos resultados posibles (exito y fracaso) y los intentos tienen un exito constante

```python
from scipy.stats import binom

# Parámetros de la distribución Binomial
# n = numero de motores
n = 1000
# p = probabilidad de que un motor sea defectuoso
p = 0.001


# 1. ¿Cuál es la probabilidad de que ninguno sea defectuoso?
x1 = 0
probabilidad_0_defectuosos = binom.pmf(x1, n, p)
print("1. La probabilidad de que no haya ninguno defectuoso es:", probabilidad_0_defectuosos)

# 2. ¿Cuál es la probabilidad de que tres sean defectuosos?
x2 = 3
probabilidad_3_defectuosos = binom.pmf(x2, n, p)
print("2. La probabilidad de que haya 3 defectuosos es:", probabilidad_3_defectuosos)

# 3. ¿Cuál es la probabilidad de que cuatro sean defectuosos?
x3 = 4
probabilidad_4_defectuosos = binom.pmf(x3, n, p)
print("3. La probabilidad de qe haya 4 defectuosos es:", probabilidad_4_defectuosos)
```
