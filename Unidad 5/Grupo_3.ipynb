```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
from sklearn.linear_model import LinearRegression
import numpy as np
from scipy.stats import pearsonr

# Cargamos el archivo del grupo 1
G3 = pd.read_csv('grupo3.csv')
G3.head()
```

```python
# Histograma de calificaciones
# Calacular la media y guardarla en una variable para mostrarla en el histograma
media = G3.Calificacion.mean()
print(media)

plt.hist(G3.Calificacion, bins = 7)
plt.title('Histograma de Calificacion')
plt.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.xlabel('Calificacion')
plt.ylabel('frecuencia')
```

```python
X = np.array(G3["HorasEstudio"]).reshape(-1, 1)
Y = G3["HorasEstudio"]

# Entrenar el modelo de regresion
modelo = LinearRegression()
modelo.fit(X, Y)

# Obtener la linea de regresion
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_


print(f"Pendiente (m): {pendiente:.2f}")
print(f"Intercepto (b): {intercepto:.2f}")
print(f"Ecuacion de regresion: Y = {pendiente:.2f} * (X) + {intercepto:.2f}")
```

```python
# Definimos tama√±o
plt.figure(figsize=(6, 4))

# Datos de entrada
Horas_Estudio = X
Calificacion = Y

# Creamos el diagrama de dispersion con la linea de regresion
sns.set(style="whitegrid")
sns.regplot(x=Horas_Estudio, y=Calificacion, scatter=True, color='blue')
plt.title('Diagrama de Dispersion y Regresion Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacion')
plt.grid(True)
plt.show()
```

```python
coeficiente_correlacion, p_valor = pearsonr(X.flatten(), Y)
print(f"Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```
